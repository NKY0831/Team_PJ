```python
class Titanic :    
    def __init__(self) :
        import numpy as np
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        import scipy.stats as stats
        import warnings
        warnings.filterwarnings('ignore')
    
        self.df = sns.load_dataset("titanic")
        self.df1 = self.df.copy()
        self.df1.isnull().sum()
    
        print(self.df.head())
        print(self.df.tail())
        print(self.df1.describe())
        print(self.df1.info())

    def pclass_survived(self, val1, val2, plot1, plot2):
        import numpy as np
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        import scipy.stats as stats
        import warnings
        warnings.filterwarnings('ignore')
        
        plt.rc("font",family="Malgun Gothic")
        f, ax = plt.subplots(1,2, figsize= (20,8))
        sns.set(font_scale = 1)
        self.df1[val1].value_counts().plot.bar(color = ["r","b","c"], 
                                         ax=ax[0], fontsize = 20)
        ax[0].set_title(plot1)
        ax[0].set_ylabel("count")
        sns.countplot(val1, hue = val2, data = self.df1, ax = ax[1])
        ax[1].set_title(plot2)
        plt.show()
    
    def chi2(self, val1, val2) :
        import numpy as np
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        import scipy.stats as stats
        import warnings
        warnings.filterwarnings('ignore')
        
        print("카이제곱 검정테스트를 이용한",val1,"와",val2,"의 상관관계 확인")
        df1_crosstable_p = pd.crosstab(self.df1[val1], self.df1[val2])
        print(stats.chi2_contingency(df1_crosstable_p))
```

```python
Ti = Titanic()
Ti.pclass_survived("pclass", "survived","객실 등급별 탑승자수",
                   "객실 등급별 생존자와 사망자 카운트")
Ti.chi2("pclass", "survived") 
```
